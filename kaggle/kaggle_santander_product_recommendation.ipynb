# 시계열 데이터 시각화
santander_product_recommendation의 데이터는 시계열 데이터

시계열 데이터를 올바르게 분석하기 위해서는 시간에 따른 변화 척도를 눈으로 분석해야 함

고객 유입 채널을 의미하는 canal_entrada 변수도 계절, 방학/입학 시즌에 따라 데이터 분포가 변할 수 있으나, 앞의 시각화 과정에서는 시계열성 변화를 전혀 포착하지 못 함

\
**대회의 목표가 금융 제품 추천인 만큼, 시계열성에 따른 24개의 제품 변수에 대한 이해가 중요함**


*   이진 변수(0/1)로 구성된 24개의 제품 변수는 고객이 해당 제품을 보유할 때 1, 보유하지 않을 때 0 
> 각 제품 변수를 다 더한 값은 제품 보유 수준을 의미



### 월별 금융 제품 보유 데이터를 누적 막대 그래프로  시각화 하기

누적 막대그래프를 사용하는 이유
> 서로 다른 제품 간 차이를 함께 시각화하기 위함

```
#날짜 데이터를 기준으로 분석하기 위해 날짜 데이터를 별도로 추출
months = trn['fecha_dato'].unique().tolist()
#제품 변수 24개를 추출
label_cols = trn.columns[24:].tolist()
```

```
label_over_time = []
for i in range(len(label_cols)):
  # 매월 각 제품의 총합을 groupby(..).agg('sum')으로 계산하여, label_sum에 저장한다.
  label_sum = trn.groupby(['fecha_dato'])[label_cols[i]].agg('sum')
  label_over_time.append(label_sum.tolist())

label_sum_over_time = []
for i in range(len(label_cols)):
  # 누적 막대그래프 시각화를 위해, n번째 제품의 총합을 1~n번째 제품의 총합으로 만든다.
  label_sum_over_time.append(np.asarray(label_over_time[i:]).sum(axis=0))

# 시각화를 위해 색깔을 지정
color_list = ['#FF0000','#FFBB00','#ABF200','#1FDA11','#0055FF','#7E41D9','#FFD9EC','#747474']

# 그림 크기를 사전에 정의
f, ax = plt.subplots(figsize=(30, 15))
for i in range(len(label_cols)):
  # 24개 제품에 대하여 Histogram을 그림
  # x축에는 월 데이터, y 축에는 누적 총합, 색깔은 8개를 번갈아 사용하며 그림의 alpha 값은 0.7로 지정한다
  sns.barplot(x=months, y=label_sum_over_time[i], color=color_list[i%8], alpha=0.7)

# 우측 상단에 legend 추가
plt.legend([plt.Rectangle((0,0),1,1, fc=color_list[i%8], edgecolor='none') for i in range(len(label_cols))], label_cols, loc=1, ncol=2, prop={'size':16})
plt.show()
```

시각화 결과를 통해, 총 제품 보유 수량이 매달 조금씩 상승함을 확인 가능
> 고객의 숫자가 꾸준히 증가하기 때문으로 보임

### 월별 금융 제품 보유 데이터를 누적 막대 그래프로  시각화 하기(2)
절댓값이 아닌 월별 상대값으로 시각화해 시각적으로 보기 쉽게 표현


```
# label_sum_over_time값을 퍼센트 단위로 변환. 월마다 최댓값으로 나누고 100을 곱해줌
label_sum_percent = (label_sum_over_time / (1.*np.asarray(label_sum_over_time).max(axis=0))) * 100

# 시각화 실행
f, ax = plt.subplots(figsize=(30, 15))
for i in range(len(label_cols)):
  sns.barplot(x=months, y=label_sum_percent[i], color=color_list[i%8], alpha=0.7)

plt.legend([plt.Rectangle((0,0),1,1, fc=color_list[i%8], edgecolor='none') for i in range(len(label_cols))], label_cols, loc=1, ncol=2, prop={'size':16})
```

시각화를 통해 'ind_cco_fin_ult1'의 보유량은 계절과 상관 없이 꾸준히 높은 것을 확인할 수 있었음
