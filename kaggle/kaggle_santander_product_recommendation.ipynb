**0~23까지의 변수 중 int64, float64의 데이터 타입을 갖는 변수를 확인**


*   ncodpers: 최솟값 15889 ~ 최댓값 1553689를 갖는 고객 고유 식별 번호
*   ind_nuevo: 최소 75%의 값이 0, 나머지가 1을 갖는 신규 고객 지표
*   indrel: 최소 75%가 1, 나머지가 99를 가지는 고객 등급 변수
*   tipodom: 모든 값이 1 - 이러한 변수는 학습에 도움이 되지 않음

  모든 값이 상수일 경우에는 변수로서 식별력X
*   cod_prov: 최솟값 1 ~ 최댓값 52를 가짐, 수치형이지만 범주형 변수로서 의미를 가지는 지방 코드 변수
*   ind_actividad_cliente: 최소 50%의 값이 0이며 나머지가 값 1을 가지는 활발성 지표
*   renta: 최솟값 1202.73 ~ 최대 28894400의 값을 가지는 전형적인 수치형 변수. 가구 총 수입을 나타냄. 









```
num_cols = [col for col in train.columns[:24] if train[col].dtype in ['int64','float64']]
train[num_cols].describe()
```

**0~23까지의 변수 중 object의 데이터 타입을 갖는 변수를 확인**

describe에서 count, unique, top, freq의 의미

*  count: 해당 변수의 유효한 데이터 개수를 의미 - count가 작을수록 결측값이 대부분임
*   unique: 해당 범주형 변수의 고유값 개수
*   top: 가장 빈도 높은 데이터가 표시
*   freq: top에서 표시된 최빈 데이터의 빈도수 의미 - 'ind_empleado'에서는 N가 전체 데이터의 대부분(13619575/13610977)을 차지함을 알 수 있음


여기서 age와 antiguedad(은행 누적 거래 기간)은 수치형으로 분류되어야 함
